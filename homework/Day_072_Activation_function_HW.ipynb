# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return x>0

x = plt.linspace(-10,10,50)
plt.plot(x,ReLU(x))
plt.plot(x,dReLU(x))
plt.grid()

plt.title('ReLU')
plt.show()
```
